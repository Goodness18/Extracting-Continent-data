To create a database using Python, you can utilize a database management system (DBMS) such as SQLite, MySQL, PostgreSQL, or others. Here's an example using SQLite, which is a lightweight, file-based DBMS that doesn't require a separate server:

Import the required module:

```python
import sqlite3
```

Establish a connection to the database. This line creates a connection to the database file named 'database.db'. If the file doesn't exist, it will be created. If it already exists, the connection will be established.

```python
conn = sqlite3.connect('database.db')
```

The cursor is used to execute SQL statements and manage the database.

```python
#create a cursor object:
cursor = conn.cursor()
```

Write a script to create a table in the database.

```python
cursor.execute('''CREATE TABLE vehicles (
    VIN TEXT,
    County TEXT,
    City TEXT,
    State TEXT,
    PostalCode TEXT,
    Year INTEGER,
    Make TEXT,
    Model TEXT,
    ElectricVehicleType TEXT,
    CleanAlternativeFuelVehicleEligibility TEXT,
    ElectricRange INTEGER,
    BaseMSRP DECIMAL(10, 2),
    LegislativeDistrict INTEGER,
    DOLVehicleID TEXT,
    VehicleLocation TEXT,
    ElectricUtility TEXT,
    CensusTract2020 TEXT
)''')
```

```python
# Execute the SQL query to aggregate the vehicles
query = "SELECT Make, Year, COUNT(*) as Count FROM vehicles GROUP BY Make, Year"
cursor.execute(query)

# Fetch all the rows of the result
results = cursor.fetchall()
```

```python
# Execute the SQL query to get the range of BaseMSRP values
query = "SELECT MIN(BaseMSRP) AS MinValue, MAX(BaseMSRP) AS MaxValue FROM vehicles"
cursor.execute(query)

# Fetch the result
result = cursor.fetchone()

# Extract the minimum and maximum values
min_value = result[0]
max_value = result[1]

# Print the range of BaseMSRP values
print(f"Range of BaseMSRP values: {min_value} - {max_value}")
```

Update the quesry to exclude values where the Base MSRP price is zero.

```python
# Execute the SQL query to get the range of BaseMSRP values
query = "SELECT MIN(BaseMSRP) AS MinValue, MAX(BaseMSRP) AS MaxValue FROM vehicles WHERE BaseMSRP > 0"
cursor.execute(query)

# Fetch the result
result = cursor.fetchone()

# Extract the minimum and maximum values
min_value = result[0]
max_value = result[1]

# Print the range of BaseMSRP values
print(f"Range of BaseMSRP values: {min_value} - {max_value}")
```

```python
# Execute the SQL query to get the count of vehicles by Make and Year
query = "SELECT Make, Year, COUNT(*) AS VehicleCount FROM vehicles GROUP BY Year, Make"
cursor.execute(query)

# Fetch all the rows of the result
results = cursor.fetchall()

# Print the count of vehicles by Make and Year
for row in results:
    make = row[0]
    model_year = row[1]
    count = row[2]
    print(f"Year: {model_year}, Make: {make}, Vehicle Count: {count}")
```
